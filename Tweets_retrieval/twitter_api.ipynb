```python
import tweepy
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
```

```python
import api_keys
import json
```

```python
auth=OAuthHandler(api_keys.consumer_key,api_keys.consumer_secret)
auth.set_access_token(api_keys.access_token,api_keys.access_secret)
api=tweepy.API(auth)
```

```python
class std_listener(StreamListener):
    def on_data(self, data):
        with open('tweets_file.txt','a') as fp:
            fp.write(data)
            
        return True

    def on_error(self, status):
        print(status)
```

```python
var=std_listener()
stream=Stream(auth,var)
stream.filter(track=['I\'m fine','I want to die','No one cares','I am sad','anxiety','suicide','I feel sad','tired','helpless','hopeless','aching','lost','worthless','useless','stupid','stuck','adrift','hurting','alone','unsure','insecure','despair','I should','stressed'],languages=['en'])
```
